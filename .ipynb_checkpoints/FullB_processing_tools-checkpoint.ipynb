```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as fft
import os.path
from scipy.ndimage import gaussian_filter
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
def full_B_loader(fname,path=''):
    filetoload = os.path.join(path,fname)
    rawdat = np.loadtxt(filetoload,skiprows=39) # this gets the B-field image 
    with open(filetoload) as myfile:
        header = [next(myfile) for x in range(39)]
    xax = 1e6*np.linspace(float(header[9][19:]),float(header[10][18:]),int(header[11][17:]))
    yax = 1e6*np.linspace(float(header[12][19:]),float(header[13][18:]),int(header[14][17:]))
    return rawdat, xax, yax

def plotter(xax,yax,data,cmax=1.,cmin=-1.,figdims = (6,6), tickvals=[-0.2,0,0.2]):
    fig,ax = plt.subplots(figsize=figdims)
    im = ax.pcolormesh((xax-xax.min())*1e3,(yax-yax.min())*1e3,data/10,cmap='bwr',vmax=cmax,vmin=cmin)
    ax.set_xlabel('X position [nm]')
    ax.set_ylabel('Y position [nm]')
    ax.set_aspect('equal')
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)

    cbar = fig.colorbar(im, cax=cax, orientation='vertical',ticks=tickvals)
    cbar.set_label('Field / mT', rotation=270)
    fig.show()
    
    return fig, ax, im, cbar

def plotter_c(xax,yax,data,cmax=1.,cmin=-1.,figdims = (6,6), tickvals=[-0.2,0,0.2]):
    fig,ax = plt.subplots(figsize=figdims)
    im = ax.contour((xax-xax.min())*1e3,(yax-yax.min())*1e3,data/10,cmap='bwr',levels=[cmin,cmax])
    ax.set_xlabel('X position [nm]')
    ax.set_ylabel('Y position [nm]')
    ax.set_aspect('equal')
    # divider = make_axes_locatable(ax)
    # cax = divider.append_axes('right', size='5%', pad=0.05)

    # cbar = fig.colorbar(im, cax=cax, orientation='vertical',ticks=tickvals)
    # cbar.set_label('Field / mT', rotation=270)
    cbar = _
    fig.show()
    
    return fig, ax, im, cbar

def full_B_smooth(Bdat,xax,yax,cmax=1.,cmin=-1.,smooth=0.0,meansub=True,ploton=False):
    if meansub:
        print(np.mean(np.mean(Bdat,0)))
        data = Bdat - np.mean(np.mean(Bdat,0))
    else:
        data = Bdat
    datasmooth = gaussian_filter(data,smooth)
    
    if ploton:
        fig, ax, im, cbar = plotter(xax,yax,datasmooth,cmax=cmax,cmin=cmin)
    else:
        fig, ax, im, cbar = _, _, _, _
        
    
    return datasmooth, fig, ax, im, cbar

def full_B_smooth_cont(Bdat,xax,yax,cmax=1.,cmin=-1.,smooth=0.0,meansub=True,ploton=False):
    if meansub:
        print(np.mean(np.mean(Bdat,0)))
        data = Bdat - np.mean(np.mean(Bdat,0))
    else:
        data = Bdat
    datasmooth = gaussian_filter(data,smooth)
    
    if ploton:
        fig, ax, im, cbar = plotter_c(xax,yax,datasmooth,cmax=cmax,cmin=cmin)
    else:
        fig, ax, im, cbar = _, _, _, _
        
    
    return datasmooth, fig, ax, im, cbar

def B_reconstruction(xax,yax,Bdata,theta,phi):
    kx1 = fft.fftshift(fft.fftfreq(xax.size,np.mean(np.diff(xax)))) + 1e-15
    ky1 = fft.fftshift(fft.fftfreq(yax.size,np.mean(np.diff(yax)))) + 1e-15
    kx, ky = np.meshgrid(kx1, ky1)
    kmat = np.sqrt(kx**2 + ky**2)

    angle = theta/180*np.pi
    phiangle = phi/180*np.pi

    Bz_FT = fft.fftshift(np.fft.fft2(Bdata))/(np.cos(angle)-1j*np.sin(angle)*(np.cos(phiangle)*(kx/kmat)+np.sin(phiangle)*(ky/kmat)))
    Bx_FT = -1*1j*kx/kmat*Bz_FT
    By_FT = -1*1j*ky/kmat*Bz_FT

    Bx = np.real(fft.ifft2(fft.fftshift(Bx_FT)))
    By = np.real(fft.ifft2(fft.fftshift(By_FT)))
    Bz = np.real(fft.ifft2(fft.fftshift(Bz_FT)))
    return Bx, By, Bz
```

```python
# Load in data
%matplotlib inline

plt.rcParams.update({'font.size': 16})

# BFO 0 deg
path = r'Y:\Users\Paul\tempdata'
fname = r'20231212-1347-36_fullB_all_autosave_QAFM_b_field_fw.dat'

raw_data, xax, yax = full_B_loader(fname,path)
Bdata,_,_,_,_ = full_B_smooth_cont(raw_data-17.8,xax,yax,smooth=0.0,cmin=-0.08,cmax=0.08,meansub=False,ploton=True)

Bx, By, Bz = B_reconstruction(xax, yax, Bdata, 57, 0.0)

fig, ax = plt.subplots(1,3,figsize=(12,6),sharex=True,sharey=True)
ax[0].pcolormesh(xax,yax,Bx-np.mean(Bx),vmin=-3,vmax=3,cmap='inferno')
ax[1].pcolormesh(xax,yax,By-np.mean(By),vmin=-3,vmax=3,cmap='inferno')
ax[2].pcolormesh(xax,yax,Bz-np.mean(Bz),vmin=-3,vmax=3,cmap='inferno')
# ax[0].set_xticks([])
# ax[0].set_yticks([])
ax[0].set_title('B$_x$')
ax[1].set_title('B$_y$')
ax[2].set_title('B$_z$')
ax[0].set_aspect('equal')
ax[1].set_aspect('equal')
ax[2].set_aspect('equal')
fig.tight_layout()
plt.show()
```
